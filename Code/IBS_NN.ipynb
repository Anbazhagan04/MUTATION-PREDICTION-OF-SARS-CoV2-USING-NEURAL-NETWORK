```python
x=df["Sequence"].values
x
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
y=df["Variant"].values
y_labeled=le.fit_transform(y)
y_labeled
```

```python
# one_hot(x[0],100)
```

```python
x_encoded= [one_hot(d, 50) for d in x]
```

```python
max_length = len(max(x_encoded,key=len))+1
max_length 
```

```python
x_padded = pad_sequences(x_encoded, maxlen=max_length, padding='post')
print(x_padded)
```

```python
embeded_vector_size = 1000
model = Sequential()
model.add(Embedding(50, embeded_vector_size, input_length=max_length,name="embedding"))
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))
```

```python
# Categorical crossentropy
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
print(model.summary())
```

```python
x,y=x_padded,y_labeled
```

```python
model.fit(x, y, epochs=5, verbose=0)
```

```python
# evaluate the model
loss, accuracy = model.evaluate(x,y)
accuracy
```
